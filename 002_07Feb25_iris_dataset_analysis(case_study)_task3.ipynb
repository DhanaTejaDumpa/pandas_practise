```python
import pandas as pd
import numpy as np
```

### importing "iris" dataset

```python
data = pd.read_csv("iris_dataset.csv")
data
```

### Q1. Rename the columns of the dataset to 'SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', and 'Species'.


```python
data.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
data
```

### Q2.
a) Set the index of the DataFrame to the 'Species' column.

b) Retrieve the first 10 rows of the DataFrame using integer-based indexing.


```python
# a) # using Species column as index

data.set_index('Species', inplace = True)
data
```

```python
# b) Retrieving first 10 rows of dataframe using integer-based indexing
data.iloc[0:10,0:5]
```

### Q3.
a)Slice the DataFrame to display rows with 'PetalWidth' greater than 2.5.

b)Slice the DataFrame to display columns 'SepalLength' and 'SepalWidth' for the first 5 rows.


```python
# a) displaying rows with petal width greater than 2.5

data[data['PetalWidth']>2.5]
```

```python
# b) displaying first 5 rows containing Sepal length and sepal width columns

data.iloc[0:5,0:2]
```

### Q4.
a) Filter the DataFrame to show only rows where 'Species' is 'versicolor'.

b) Filter the DataFrame to show rows where 'PetalLength' is greater than 4 and 'SepalWidth' is less than 3.


```python
# a)showing only rows where species is vercicolor
data[data.index == 'Iris-versicolor']
```

```python
# b) showing rows where petal length is greater than 4 and sepal width is less than 3

data[(data['PetalLength']>4) & (data['SepalWidth']<3)]
```

### Q5.
a)Calculate the minimum 'PetalLength' across all species.

b)Calculate the maximum 'SepalWidth' across all species.


```python
# a) calculating the minimum petal length
data['PetalLength'].min()
```

```python
# b) calculating the maximum sepal width

data['SepalWidth'].max()
```

### Q6.
a) Calculate the sum of values along axis=1 (row-wise).

b) Calculate the mean of values along axis=0 (column-wise).

### Q11.
a) Find the top 3 rows with the largest 'PetalLength'.

b) Find the top 3 rows with the smallest 'SepalWidth'.


```python
# a) Finding the top 3 rows with the largest 'PetalLength'
data['PetalLength'].sort_values(ascending = False).head(3)
```

### Q12.
a) Sort the DataFrame by 'PetalWidth' in ascending order.

b) Sort the DataFrame by 'SepalLength' in descending order.


```python
# a) Sorting the dataframe by petal width in ascending order

data.sort_values(by = 'PetalWidth')
```

```python
# b) sorting the dataframe by sepal length in descending order
data.sort_values(by = 'SepalLength',ascending = False)
```
