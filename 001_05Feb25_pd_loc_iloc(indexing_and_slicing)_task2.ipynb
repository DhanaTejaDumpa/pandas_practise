### 5.Using loc, extract all rows where the gender is "Male"

### 6.Using loc, extract the chol and hdl columns for rows where age is greater than 50.

```python
# extracting chol and hd1 columns for rows where age is greater then 50.

data.loc[data['age']>50,['chol','hdl']]
```

### 7.Using iloc, extract all rows and only the columns at positions 0, 5, and 10.

```python
# extracting all rows and only the columns at positions 0,5, and 10

data.iloc[:,[0,5,10]]
```
