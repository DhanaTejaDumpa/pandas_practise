###### 19. Write a Pandas program to read the diamonds DataFrame and detect duplicate color.

```python
#### code here
duplicate_colors = diamond['color'].duplicated(keep=False)
unique_duplicate_colors = diamond.loc[duplicate_colors, 'color'].unique()
print("Duplicate Colors:", unique_duplicate_colors)
```
