### 1) Create a Pandas Series from a list and perform the following tasks:
 * print the entire series
 * print the first 3 elements of the series
 * print the last 2 elements of the series.

```python
# creating a pandas series

lst = [1,2,3,4,5,6,7,8,9]

import pandas as pd

series = pd.Series(lst)
series
```

```python
# printing first 3 elements of the series
# using .loc we access explicit indexes

series.loc[0:2]    # upperbound is inclusive
```

```python
# printing last 2 elements of the series
#using .iloc we access implicit indexes

series.iloc[7:9]    # upperbound is exclusive
```

### 2) Create two Pandas series with different data type(numeric and string) and perform the following tasks:
* Concatenate the two series into a single series.
* calculate the mean of the numeric series.

```python
# creating the numeric series

numeric_series = pd.Series([1,2,3,4,5])
numeric_series
```

```python
# creating the string series

string_series = pd.Series(['a','b','c','d','e'])
string_series
```

```python
# conactenating numeric series and string series

# numeric_series + string_series                 #uncomment to see the error

```

## Note :             
### Explanation :              
* performing addition operation between a series containing inetegers(numeric_series) and a series containing strings(string_series) raises a "Type Error".As we can't use `+` operator for two different data types and cannot be directly combined .

* Instead we can use the <mark>pd.concat</mark> function from the pandas library.
* pd.concat is designed to combine Series or DataFrames along a particular axis. By default, it concatenates along axis 0 (rows), effectively stacking the two series on top of each other into a single series. This avoids the type error because it is simply combining the series, not attempting to perform any arithmetic operation between the elements.

```python
# concatenating both series using pd.concat

# syntax : pd.concat([series1,series2])

concatenated_series = pd.concat([numeric_series,string_series])
concatenated_series      # concatenated along axis 0(rows)...stacking two series on top of each other
```

```python
concatenated_series = pd.concat([numeric_series,string_series],axis = 1)
concatenated_series      # concatenated along axis 1(columns)...stacking two series side by side
```

```python
# calculating the mean of numeric series
# using mean() method

mean = numeric_series.mean()
mean
```

### 3) Create a Pandas Series with data from a dictionary and perform the following tasks:
* print the series
* check if a specific index exists in the series
* Add a new eleemnt to the series with a new index

```python
# creating series using dictionary
# printing the series

dict = {'a' : 1, 'b' : 2, 'c' : 3, 'd' : 4, 'e' : 5}

series = pd.Series(dict)   # keys becomes indexes of the series
series                     # values becomes data of the series
```

```python
# checking if a specific index exists in the series
'a' in series.index     # using membership operator
```

```python
# Adding a new element

new_element = pd.Series({'f' : 6})    # here f is new index and 6 is new data value
series = pd.concat([series,new_element]) # updating the series using concatenation
series
```

```python
# Alternative method

series['g'] = 7    # direct assignment
series
```

**Note :**
* pandas series are similar to dictionaries
* you can directly assign a value to a new index in the series.
* pandas will automatically add this new index and value to the series.
